5-Hangi ülkedekiler daha fazla bira içiyor ortalamada gösteriniz.

```python
# Here is list of countiries which drinks beer more than total average and sorted rows for seeing most to lowest one

avg = df['beer_servings'].mean()
df_ = df.sort_values(by = ['beer_servings'],ascending=False)
df_['country'][df_['beer_servings']>avg]
```

6-Her kıta için şarap tüketiminin betimsel istatiğini yapınız.

```python
## For Comparing wine_servings values of continent 
# First is decribe of our table
# Second is values of average , max , standard deviation , sum of all values

ag = df.groupby(['continent']).agg({'wine_servings' : [np.mean,max,np.std,sum]})
print(ag.describe())
ag
```

7-Her kıta için ve her alkol cinsi için ortalama alkol tüketimini gösteriniz. 

```python
# Used group by for make group each continent
df.groupby('continent').mean()
```

8-Her kıta için ve aşağıdaki alkol cinsleri için belirtilen hesaplamaları gerçekleştiriniz.

beer_servings - std
spirit_servings - median 
wine_servings - min, mean, max
total_litres_of_pure_alcohol - mean, count

```python
# Imported numpy for median and mean functions adn used aggleration for sperating

df.groupby('continent').agg({
        'beer_servings':[np.std],
        'spirit_servings':[np.median],
        'wine_servings' : [min,np.mean,max],
        'total_litres_of_pure_alcohol' : [np.mean,max]
})
```

9- Her alkol cinsinin en çok tüketildiği ülkeyi bulun. 

```python
## Will be tried to fix better solution
# Little fix (But can be better ofc)
# Find most consumes values and make new dataframe with them
a = []
print(df['country'].loc[df['beer_servings'].idxmax()] + ' consumes the most beer')
a.append(df.loc[df['beer_servings'].idxmax()])
print(df['country'].loc[df['spirit_servings'].idxmax()] + ' consumes the most spirit')
a.append(df.loc[df['spirit_servings'].idxmax()])
print(df['country'].loc[df['wine_servings'].idxmax()] + ' consumes the most wine')
a.append(df.loc[df['wine_servings'].idxmax()])
cntry = pd.DataFrame(a)
cntry
```

10- Yeni değişkenler üreterek her alkol cincisin ilgili ülkedeki toplam alkol tüketimi içerisindeki oranlarını yazdırınız.  

```python
## There is NaN values because in some countries total_litres_of_pure_alcohol value is zero so it can't divided by 0


df['beer/total_alcohil'] = df.beer_servings/df.total_litres_of_pure_alcohol
df['sprit/total_alcohil'] = df.spirit_servings/df.total_litres_of_pure_alcohol
df['wine/total_alcohil'] = df.wine_servings/df.total_litres_of_pure_alcohol
df
```

1- Persona.csv verisini kullanarak EDA (Exploratory data analysis) yapın.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', 170)
```

```python
df = pd.read_csv('persona.csv')

def EDA(df):
    print(str(df.shape) + '\n')
    print(str(df.info()) + '\n')
    print(str(df.describe().T) + '\n')
    print(str(df.isnull().sum()) + '\n')
    print(str(df.isnull().sum()) + '\n')
    print(df.dtypes)

EDA(df)
```

2-Persona.csv verisini kullanarak Kategorik ve Numerik Değişkenleri belirleyin.

```python
Categoric =[]
Numeric = []
for col in df.columns:
    if df[col].dtypes == "O":
        Categoric.append(col)
    else:
        Numeric.append(col)
Categoric
```

3-Persona.csv verisini kullanarak Kategorik ve Numerik Değişkenleri analiz edin. Yorumda bulunun.

```python
## Here is Numeric variables graph for its values changes
## For age is mosyly unde 20 years. 
for i in Numeric:
    df[i].hist(bins=30)
    plt.title(i)
    plt.show()
```

```python
for i in Categoric:
    sns.countplot(x=df[i], data=df)
    plt.title(i)
    plt.show()

## Users generally has Android but counts are similar approximately %60-%40
## About Sex is same seperated equally like %45 male %55  female
##about country mosly from USA after bra
```
